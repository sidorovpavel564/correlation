```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from sklearn.datasets import make_blobs
mpl.rcParams['figure.dpi'] = 300
```

```python
# нелинейная корреляция
data_columns = ['column1', 'column2']
rng = np.random.default_rng(12345)
column_a = list(rng.integers(low=0, high=100, size=100))
column_b = [x ** 2 for x in column_a]
result = zip(column_a, column_b)
data = pd.DataFrame(result, columns=data_columns)
data
```

```python
data.corr(method='spearman')
```

```python
scatter_kws = {'s': 20}
line_kws = {'color': 'red', 'lw': 1}
sns.lmplot(data=data, x='column1', y='column2', scatter_kws=scatter_kws, line_kws=line_kws, order=2)
plt.show()
```
