```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from sklearn.datasets import make_blobs
mpl.rcParams['figure.dpi'] = 300
```

```python
# нулевая корреляция
data_columns = ['column1', 'column2']
rng = np.random.default_rng(12345)
column_a = list(rng.integers(low=0, high=100, size=100))
rng = np.random.default_rng(54321)
column_b = list(rng.integers(low=0, high=100, size=100))
result = zip(column_a, column_b)
data = pd.DataFrame(result, columns=data_columns)
data
```

```python
data.corr()
```

```python
scatter_kws = {'s': 20}
line_kws = {'color': 'red', 'lw': 1}
sns.regplot(data=data, x='column1', y='column2', scatter_kws=scatter_kws, line_kws=line_kws)
plt.show()
```
