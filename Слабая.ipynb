```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from sklearn.datasets import make_blobs
mpl.rcParams['figure.dpi'] = 300
```

```python
# сильная корреляция
data_columns = ['column1', 'column2']
column_a = [10, 15, 17, 23, 26, 27, 33, 35, 37, 42, 47, 54, 60, 65, 67, 70, 77, 89, 97, 100]
column_b = [55, 1, 7, 77, 15, 65, 30, 77, 20, 46, 30, 1, 25, 63, 20, 75, 12, 90, 77, 87]
result = zip(column_a, column_b)
data = pd.DataFrame(result, columns=data_columns)
data
```

```python
data.corr()
```

```python
scatter_kws = {'s': 20}
line_kws = {'color': 'red', 'lw': 1}
sns.regplot(data=data, x='column1', y='column2', scatter_kws=scatter_kws, line_kws=line_kws)
plt.show()
```
